# RB-2-1 Real World Programming

Object-Oriented Programming (OOP) adalah paradigma pemrograman yang menggabungkan attribute (property) dan behavior yang saling terkait dalam sebuah object.

Daengan kata lain, OOP adalah pendekatan untuk memodelkan object dunia nyata menjadi object dalam program komputer yang memiliki attribut dan dapat menjalankan method tertentu.

Istilah dalam OOP:
- `oject` adalah hal apapun yang datanya ingin kita simpan dan proses.
- `class` adalah code yang mendefinisikan attribute dan behavior dari sebuah object (*bluprint* untuk membuat object).
- `attribute` (*property*) adalah variable yang terdapat di dalam sebuah class. Attribute umumnya berfungsi untuk menyimpan data sebuah instance.
- `method` adalah function yang terdapat di dalam sebuah class.
- `instance` adalah object yang memiliki nilai attribute
- `instantiation` adalah pemberian nilai attribute pada sebuah object (proses pembuatan instance)

# Konsep Dasar OOP

## Encapsulation

Pada program sebelumnya, kita melalukan instantiation secara terpisah dari pembuatan object.

Kali ini, kita akan memodifikasi program sebelumnya sehingga kita bisa melakukan instantiation bersamaan dengan pembuatan object dengan menambahkan `constructor`. `Constructor` ditandai dengan method bernama `__init__()`.

Selain itu, kita juga akan menambahkan method setter dan getter pada class kucing dan sedikit penyesuaian pada method lainnya.

constructorcat (1).png

```python
class kucing:
  #constructor
  def __init__(self, nama, jenis_kelamin, ras, umur, warna):
    self.nama = nama
    self.jenis_kelamin = jenis_kelamin
    self.ras = ras
    self.umur = umur
    self.warna = warna
  
  #metod getter
  def get_info_kucing(self):
    return self.nama,  self.jenis_kelamin, self.ras,  self.umur, self.warna

  #method setter
  def set_info_kucing(self, nama, jenis_kelamin, ras, umur, warna):
    self.nama = nama
    self.jenis_kelamin = jenis_kelamin
    self.ras = ras
    self.umur = umur
    self.warna = warna
  
  def bersuara(self):
    print(f'{self.nama} bersuara "meow"')
  
  def tidur(self):
    print(f'{self.nama} sedang tidur')
```

## Inheritance

Contoh implementasi OOP yang menerapkan konsep *inheritance*:

Basic Wireframe (16).png

orang adalah *parent class*, sedangkan dosen dan mahasiswa adalah *child class*.

Sebagai parent class, class orang mewariskan 3 attribute (nama, jenkel, dan alamat) dan 1 method (`get_info()`) kepada class dosen dan mahasiswa. Attribute dan method yang diwariskan bisa digunakan oleh child class.

Selain mendapat warisan attribute dan method dari class orang, class dosen dan mahasiswa juga memiliki attribute dan method-nya sendiri. Hal ini berarti class dosen dan mahasiswa telah memperluas (extend) attribute dan method.

Program dari class diagram diatas adalah sebagai berikut:

kampus (2).png

```python
#parent class
class orang:
  nama = ''
  jankel = ''
  alamat = ''

  def get_info(self):
    return self.nama, self.jenkel, self.alamat

#child class 1
class dosen(orang):
  nidn = ''

  def input_nilai(self, nilai):
    return nilai

#child class 2
class mahasiswa(orang):
  nim = ''
  nilai = 0

  def set_nilai(self, nilai):
    self.nilai = int(nilai)

  def get_nilai(self):
    return self.nilai
```

Membuat instance dari masing-masing child class:

kampus2 (1).png



```python
#membuat object class dosen dan instantiation
wibowo = dosen()
wibowo.nama = 'Wibowo'
wibowo.jenkel = 'Pria'
wibowo.alamat = 'Bandung'
wibowo.nidn = '202781'

#membuat object class mahasiswa dan instantiation
anita = mahasiswa()
anita.nama = 'Anita'
anita.jenkel = 'Wanita'
anita.alamat = 'Bogor'
anita.nidn = '02001747000'
```

Memanggil method warisan melalui instance child class:

## Polymorphism

Contoh implementasi OOP yang menerapkan *polymorphism*:

Basic Wireframe (17).png

class persegi dan lingkaran mendapat warisan method `hitung_luas()` dari class bangun_datar.

Fungsi dari method `hitung_luas()` adalah untuk menghitung luas bangun datar. Karena bangun datar persegi dan lingkaran memiliki rumus luas yang berbeda, maka isi method `hitung_luas()` pada kedua child class juga akan berbeda. Inilah yang dinamakan konsep *polymorphism*, yakni method bernama sama namun dengan isi/bentuk yang berbeda.

Program dari class diagram di atas adalah sebagai berikut:

polymorphism1 (2).png

```python
#parent class
class bangun_datar:
  luas = 0

  def hitung_luas(self):
    print('Hitung Luas Bangun Datar')
  
  def get_luas(self):
    return self.luas

#child class 1
class persegi(bangun_datar):
  def __init__(self, sisi):
    self.sisi = sisi
  
  def hitung_luas(self):
    self.luas = self.sisi * self.sisi

#child class 2
class lingkaran(bangun_datar):
  def __init__(self, jari2):
    self.jari2 = jari2

  def hitung_luas(self):
    self.luas = 3.14 * self.jari2**2
```

Mari kita buat object atau instance pada tiap class kemudian jalankan method `hitung_luas()` melalui object atau instance tersebut:

- Class bangun_datar:

parent.PNG

```python
tes = bangun_datar()
tes.hitung_luas()
tes.get_luas()
```

- Class persegi

persegi.PNG

```python
P1 = persegi(10)
P1.hitung_luas()
P1.get_luas()
```

- Class lingkaran

lingkaran.PNG

```python
L1 = lingkaran(10)
L1.hitung_luas()
L1.get_luas()
```

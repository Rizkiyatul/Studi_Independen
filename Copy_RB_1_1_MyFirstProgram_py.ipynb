Masalah ini dapat kita tangani dengan memanfaatkan *Type Casting*.

*Type Casting* adalah metode untuk mengkonversi tipe data dalam suatu variable menjadi tipe data lain. Tipe data hasil konversi selanjutnya dapat kita manipulasi sesuai kebutuhan kita.

Type Casting dapat dilakukan melalui data type function, antara lain:

<br>

Data Type Function | Deskripsi
-------------------|-------------------

`float()`| Mengkonversi tipe data menjadi integer
`str()` | Mengkonversi tipe data menjadi string (str)
`bool()` | Mengkonversi tipe data menjadi boolean (bool)

# Operator

Operator adalah simbol-simbol yang dapat kita gunakan untuk memanipulasi data atau nilai dalam variable. Secara umum, terdapat 3 jenis operator, yakni:

- Operator aritmatika
- Operator perbandingan
- Operator logika

## Operator Aritmatika

Operator aritmatika digunakan untuk melakukan operasi matematika pada data atau nilai variable. Berikut ini adalah operator aritmatika:

<br>

Operator | Deskripsi
---|---
`**` | Eksponen (Pemangkatan)
`*` | Perkalian
`/` | Pembagian
`//` | Floor division
`%` | Modulus
`+` | Penambahan
`-` | Pengurangan

<br>
<br>

aritmatika.PNG

```python
x = 5
y = 3

exponent = x ** y
multiplication = x * y
division = x / y
floor_division = x // y
modulus = x % y
addition = x + y
subtraction = x - y

print(f'{x} ** {y} = {exponent}') # x pangkat y
print(f'{x} * {y}  = {multiplication}')
print(f'{x} / {y}  = {division}')
print(f'{x} // {y} = {floor_division}') # x pembagian dnegan pembulatan ke bawah
print(f'{x} % {y}  = {modulus}') # sisa hasil bagi x dengan y 
print(f'{x} + {y}  =  {addition}')
print(f'{x} - {y}  = {exponent}')
```

## Operator Perbandingan

Sesuai dengan namanya, operator perbandingan digunakan untuk membandingkan dua nilai. Output operator ini adalah nilai boolean. Berikut ini adalah operator perbandingan:

<br>

Operator | Deskripsi
-|-
`<`| Kurang dari
`>`| Lebih dari
`<=`| Kurang dari sama dengan
`>=`| Lebih dari sama dengan
`==`| Sama dengan
`!=`| Tidak sama dengan

<br>
<br>

comparison.PNG

```python
x = 5
y = 3

print(f"{x} < {y} =", x < y)
print(f"{x} > {y} =", x > y)
print(f"{x} <= {y} =", x <= y)
print(f"{x} >= {y} =", x <= y)
print(f"{x} == {y} =", x != y)
print(f"{x} != {y} =", x == y)
```

Operator perbandingan juga bisa digunakan untuk membandingkan string.

string compare.PNG

```python
password = 'admin123'
password_tersimpan = 'admin123'

verifikasi = password == password_tersimpan

print(verifikasi)
```

## Operator Logika

Operator logika digunakan untuk mengevaluasi dua atau lebih kondisi. Output operator logika juga berupa nilai boolean.

Berikut adalah operator logika:
- `and`
- `or`
- `not`


logic.PNG

```python
x = True
y = False

print("x and b = ", x and y)
print("x or y  = ", x or y)
print("not(x)  = ", not(x))
```

Operator ini juga bisa dipakai untuk menghubungkan dua atau lebih operator perbandingan.


logiccomparison.PNG

```python
a = 9
b = 12
c = 35

print(a < b and b > c)
print(a == b and b != c, '\n')

print(a < b or b < c)
print(a == b and b !=c, '\n')

print(not(a == b))
print(not(b != c))
```
